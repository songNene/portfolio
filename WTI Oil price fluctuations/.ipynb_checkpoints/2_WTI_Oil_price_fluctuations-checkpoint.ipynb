```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

## WTI 가격 VS 달러지수

```python
# WTI 가격 추세 확인하기
plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['WtiPrice'])
plt.title('WTI Weekly Price Trend')
plt.xlabel('Date')
plt.ylabel('WTI Price')
plt.show()
```

**결과 :**    

WTI 주간 가격은 장기적인 상승·하락 국면은 존재하지만 일정한 주기성을 반복하는 패턴은 뚜렷하게 관찰되지 않는다.    
특히 일부 시점에서는 외생 충격에 의해 단기간 급격한 하락 또는 급등이 발생하며 이러한 구간이 전체 변동성을 지배하는 구조를 보인다.     
계절성이나 고정된 주기보다는 이벤트 중심의 비정기적 변동이 더 두드러진다.

```python
# 달러지수 추세 확인
plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['UsdIndex'])
plt.title('Dollar Index Weekly Trend')
plt.xlabel('Date')
plt.ylabel('Dollar Index')
plt.show()
```

**결과 :** 

달러지수는 WTI 가격과 달리 급격한 상승 국면 이후 비교적 완만한 하락이 이어지는 패턴이 반복적으로 관찰된다.    
달러지수는 통화 정책과 같은 거시 금융 요인의 영향을 받으며 이러한 요인은 원자재 시장에도 구조적인 영향을 미치는 것으로 알려져 있다.

```python
# WTI 가격 VS 달러지수

# 정규화(스케일링)
scaled = pd.DataFrame(
    StandardScaler().fit_transform(df[["WtiPrice", "UsdIndex"]]),
    columns=["WTI_scaled", "Dollar_scaled"]
)

# 시각화
plt.figure(figsize=(12, 4))
plt.plot(df["Date"], scaled["WTI_scaled"], label="WTI (scaled)")
plt.plot(df["Date"], scaled["Dollar_scaled"], label="Dollar Index (scaled)")
plt.legend()
plt.title("WTI vs Dollar Index (Scaled Trend)")
plt.show()
```

**결과:**    
비교한 결과 두 지표 간에는 장기적인 동조보다는 시기별로 관계가 달라지는 양상이 확인되었다.     
이는 달러지수가 항상 즉각적으로 WTI 가격을 설명하지는 않으며 다른 외생 변수와 함께 해석할 필요가 있음을 보여준다.


```python
# 변화량 확인하기 하기
df['wti_return'] = df['WtiPrice'].pct_change()
df['usd_return'] = df['UsdIndex'].pct_change()
```

```python
# 변화량 시각화 
plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['wti_return'], label='WTI return')
plt.plot(df['Date'], df['usd_return'], label='USD return')
plt.legend()
plt.title('WTI vs Dollar Index Returns')
plt.show()
```

**결과:**    
가격 수준이 아닌 변화율 기준으로 두 지표를 비교한 결과 특정 시점에서 단기 충격이 동시에 혹은 비대칭적으로 발생하는 구간이 관찰되었다.     
특히 WTI 수익률은 일부 주차에서 매우 큰 스파이크를 보이며 이러한 급변 구간이 리스크 탐지의 핵심 대상이 됨을 확인하였다.     
이를 통해 단순 가격 예측이 아닌 변동성 확대 구간을 사전에 감지하는 접근의 필요성을 확인하였다.


```python
# 몇 주 뒤 영향 확인 (과거 달러 움직임이 이후 WTI 변화와 얼마나 같이 움직였는가의 지표)
for lag in [1, 2, 4, 8]:
    corr = df['wti_return'].corr(df['usd_return'].shift(lag))
    print(f'Lag {lag} week correlation: {corr:.3f}')
```

**결과:** 

WTI와 달러지수의 주간 변화율을 비교한 결과 전체 기간 평균 상관계수는 낮게 나타났으며 두 변수 간의 관계는 시계열 전반에 걸쳐 일관되지 않았다.    
다만, 일부에서 두 변수의 변동이 동시에 확대되는 구간이 관찰되었으나 이는 달러지수의 영향이 특정 국면에서만 강화될 가능성을 시사한다.

```python
# 4주 이동 변동성 계산
# 주간 수익률 기준으로 4주(1개월) 이동 표준편차를 계산하여 단기 가격 변동성 수준을 정량화한다.
# → 변동성 급증 여부를 판단하기 위한 기초 지표

df['wti_vol_4w'] = df['wti_return'].rolling(4).std()
df['usd_vol_4w'] = df['usd_return'].rolling(4).std()
```

```python
# WTI vs USD 변동성 구조 비교 시각화 

plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['wti_vol_4w'], label='WTI 4W Vol')
plt.plot(df['Date'], df['usd_vol_4w'], label='USD 4W Vol')
plt.legend()
plt.title('WTI vs USD Rolling Volatility (4 Weeks)')
plt.show()
```

**결과:**     
달러지수의 변동성은 전 기간에 걸쳐 상대적으로 안정적인 수준을 유지한 반면 WTI 변동성은 특정 시점에서 매우 급격한 스파이크 형태로 집중되어 나타났다.    
특히 2020년 전후 구간에서 WTI 변동성이 단기간에 폭발적으로 확대되었으며 이는 원유 시장이 거시 금융 변수보다 지정학적·수급·정책 이벤트에 훨씬 민감하게 반응함을 시사한다.

이러한 결과는 원유 가격 리스크가 지속적인 변동성 확대보다는 특정 시점에 집중적으로 발생하는 위기형 변동성 구조를 가진다는 점을 보여준다.    
따라서 평균적인 변동성 예측보다, 변동성 급증 구간을 사전에 탐지하는 접근이 더 합리적임을 확인하였다.

```python
# 고변동성(위기) 구간 정의 - 4주 이동 변동성(wti_vol_4w)의 상위 10% 구간
threshold = df['wti_vol_4w'].quantile(0.9)
df['high_vol'] = (df['wti_vol_4w'] > threshold).astype(int)
```

```python
# 고변동성 구간 시각화
# 전체 WTI 가격 흐름 위에 변동성 급증 시점만 시각적으로 강조하여 가격 급등락과 위기 구간의 대응 관계를 확인
import matplotlib.dates as mdates

plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['WtiPrice'], alpha=0.6)

plt.scatter(
    df.loc[df['high_vol'] == 1, 'Date'],
    df.loc[df['high_vol'] == 1, 'WtiPrice'],
    color='red',
    s=10,
    label='High Volatility'
)

plt.legend()
plt.title('WTI Price with High Volatility Periods')
plt.show()
```

**결과:**    
고변동성 구간은 가격이 안정적으로 움직이는 구간 전반에 분포하기보다는 특정 시점에 집중적으로 나타나는 양상이 확인되었다. 

특히 가격이 급격히 하락하거나 단기간에 방향성이 급변하는 국면에서 고변동성 신호가 밀집되어 나타났으며    
이는 단순한 가격 수준 변화보다 가격 움직임의 불안정성이 위기 구간을 설명하는 데 더 중요한 역할을 함을 시사한다.

이러한 결과는 WTI 가격 리스크가 점진적으로 누적되기보다는 외생 이벤트나 구조적 충격에 의해 단기간에 집중적으로 발생하는 특성을 가진다는 점을 보여주며    
평균적인 가격 예측보다는 변동성 급증 구간을 사전에 식별하는 접근이 리스크 관리 측면에서 더 적합함을 확인하는 근거가 된다.

```python
# 고변동성 달러지수 확인하기
df.groupby('high_vol')['UsdIndex'].describe()
```

**결과 :**     
WTI 고변동성 구간과 일반 구간을 기준으로 달러지수의 분포를 비교한 결과 두 구간 간 달러지수의 평균 수준 차이는 크지 않게 나타났다.     
이는 원유 시장의 고변동성 국면이 달러지수의 절대적인 상승 또는 하락 국면과 직접적으로 일치하지는 않음을 의미한다.

다만 고변동성 구간에서는 달러지수의 표준편차와 범위가 소폭 증가하는 경향이 관찰되었다.    
이는 원유 변동성이 확대되는 시점에 달러지수 역시 보다 불안정한 움직임을 보이며 글로벌 금융 환경 전반의 불확실성이 함께 증가했음을 시사한다.

```python
# 고변동성 구간에서의 달러지수 변동성
df.groupby('high_vol')['usd_vol_4w'].mean()
```

**결과:**     
달러지수의 4주 이동 변동성 평균을 비교한 결과 고변동성 구간에서 달러지수 변동성이 일반 구간보다 다소 높게 나타났다.     
WTI 변동성 확대가 달러지수 변동성과 완전히 독립적으로 발생하기보다는 글로벌 금융 시장 전반의 변동성 국면과 부분적으로 동조하는 구조를 가진다는 점을 보여준다.

달러지수는 WTI 고변동성의 직접적인 원인 변수라기보다는 원유 시장 리스크가 확대되는 거시적 환경을 설명하는 보조적인 외생 변수로 해석하는 것이 적절하다.

# WTI 가격 VS 미국 10년 국채 금리

```python
# 추세 확인하기
plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['US_10yRate'])
plt.title('US 10Y Treasury Yield (Weekly)')
plt.xlabel('Date')
plt.ylabel('Rate (%)')
plt.show()
```

**결과:**    
미국 10년물 국채 금리는 분석 기간 동안 몇 차례의 뚜렷한 국면 전환을 보이며 2020년 이후 장기 저금리 국면에서 고금리 국면으로 이동하는 변화가 관찰된다.    
이러한 금리의 국면적 변화는 금융 시장 전반의 유동성 환경을 반영하며 원유 시장에서도 변동성 확대가 나타나는 배경 조건으로 작용할 수 있다.

```python
# 변화량 확인 (차분 -diff 사용)
df['rate_change'] = df['US_10yRate'].diff()
```

```python
# 시각화
plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['rate_change'], label='10Y Rate Change')
plt.axhline(0, linewidth=0.8)
plt.legend()
plt.title('US 10Y Rate Weekly Change')
plt.show()
```

**결과:**    
미국 10년물 국채 금리의 주간 변화량은 대부분의 기간 동안 평균 0을 중심으로 제한된 범위에서 움직이지만 일부 시점에서는 급격한 변동이 집중적으로 나타난다.    
이러한 급격한 금리 변화 구간은 금융 시장 전반의 불확실성이 확대된 시점으로 해석할 수 있으며 유가 변동성 확대와의 관계를 추가적으로 탐색할 필용성을 시사한다.

```python
# 최근 4주 변동량 계산
df['rate_vol_4w'] = df['rate_change'].rolling(4).std()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['wti_vol_4w'], label='WTI 4W Vol')
plt.plot(df['Date'], df['rate_vol_4w'], label='10Y Rate 4W Vol')
plt.legend()
plt.title('WTI vs US 10Y Rate Rolling Volatility (4 Weeks)')
plt.show()
```

**결과:**    
WTI와 미국 10년물 금리의 변동성을 비교한 결과 일부 극단적 위기 구간에서는 두 변수의 변동성이 동시에 확대되는 현상이 관찰되었다.    
하지만 이러한 동조 현상은 전 기간에 걸쳐 일관되게 나타나지는 않으며, 금리 변동성은 유가 변동성의 단독 설명 변수로 보기에는 한계가 있다.

```python
# 고변동성 조건에서 금리 상태 비교
df.groupby('high_vol')['US_10yRate'].describe()
```

```python
df.groupby('high_vol')['rate_vol_4w'].mean()
```

**결과 :**    
WTI 고변동성 구간을 기준으로 금리 상태를 비교한 결과 금리의 절대적 레벨은     
유가 변동성 확대를 설명하지 못하는 반면 금리 변화량의 변동성은 고변동성 구간에서 평균적으로 더 높게 나타났다.

이는 유가 리스크가 특정 금리 수준보다는 금융 환경의 불안정성과 더 밀접하게 연관되어 있을 가능성을 시사한다.

# WTI 가격 VS S&P 500

```python
# 추세 확인하기
plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['Sp500Index'])
plt.title('S&P500 Index (Weekly)')
plt.xlabel('Date')
plt.ylabel('Index Level')
plt.show()
```

**결과:**    
S&P500 지수는 분석 기간 전반에 걸쳐 장기적인 우상향 추세를 유지하고 있으며 이는 위험자산 시장의 구조적 성장 흐름을 반영한다.    
다만 2020년과 2025년에는 단기간에 비교적 큰 폭의 하락이 관찰되며 2022년 후반부터 2023년까지는 급락보다는 완만한 조정 국면이 이어지는 특징을 보인다.    
이러한 하락 구간은 장기 추세의 붕괴라기보다는 시장 불확실성이 확대된 국면에서의 단기적 가격 조정으로 해석할 수 있다

```python
# 수익률 계산
df['sp500_return'] = df['Sp500Index'].pct_change()
```

```python
# S&P500 주간 수익률 시계열 확인
plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['sp500_return'], label='S&P500 Return')
plt.axhline(0, linewidth=0.8)
plt.legend()
plt.title('S&P500 Weekly Return')
plt.show()
```

**결과:**    
S&P500 주간 수익률은 대부분의 기간 동안 0을 중심으로 비교적 안정적인 분포를 보이지만 일부 구간에서는 급격한 변동이 집중적으로 나타난다.    
특히 2020년에는 단기간에 큰 폭의 양·음 수익률이 반복적으로 관찰되며 시장 변동성이 극단적으로 확대된 모습을 보인다.    
반면 2022년에는 2020년과 같은 단발성 충격보다는 비교적 큰 음의 수익률이 여러 차례 지속적으로 발생하며 장기간에 걸친 조정 국면의 특성이 나타난다.

```python
# 최근 4주 기준 변화량
df['sp500_vol_4w'] = df['sp500_return'].rolling(4).std()
```

# Part 2. 고변동성 구간(high_vol) 분류 모델링

“다음 주(또는 현재 주)에 WTI 시장이 고변동성 국면(high_vol)에 진입할지를 거시·금융·이벤트 신호를 이용해 분류할 수 있는가?”

## Logistic Regression

```python
# X / y 데이터 만들기
feature_cols = [
    # WTI 단기 충
    'wti_return',

    # 금융 긴장도
    'usd_return',
    'usd_vol_4w',
    'rate_change',
    'rate_vol_4w',

    # 주식시장 리스크
    'sp500_return',
    'sp500_vol_4w',

    # 수급 관련 충격
    'inventory_change',
    'inventory_vol_4w',

    # 정책 및 이벤트
    'OpecEventDummy',
    'OPEC_ProdDirection'
]

X = df[feature_cols].copy()
y = df['high_vol']
```

```python
# 결측치 처리: 이벤트 없음/초기 rolling 구간 결측을 베이스라인에서는 0으로 단순 대체
# 베이스라인에서는 0으로 단순 대체, 이후에는 ffill/마스크 변수 등으로 개선 가능
X = X.fillna(0)
```

```python
# 시계열 분류: 시간 순서 유지(셔플 금지), 2023년 이후를 검증 구간으로 분리
# 미래 정보 누수를 막기 위해 시간 기준 분리
split_date = '2023-01-01'

X_train = X[X.index < df[df['Date'] >= split_date].index[0]]
X_valid = X[X.index >= df[df['Date'] >= split_date].index[0]]

y_train = y.loc[X_train.index]
y_valid = y.loc[X_valid.index]
```

```python
# 베이스라인 분류 모델 (Logistic Regression)
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score

# 베이스라인: 스케일링 없이 기본 Logistic Regression으로 동작 확인
# 이후 단계에서 StandardScaler + class_weight/threshold 튜닝 적용

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)

pred_proba = lr.predict_proba(X_valid)[:, 1]
pred_label = (pred_proba > 0.5).astype(int)

print("== Logistic Regression ==")
print(classification_report(y_valid, pred_label))
print("=" * 20)
print('ROC AUC:', roc_auc_score(y_valid, pred_proba))
```

**결과:**    
모델이 검증 구간에서 high_vol(1)을 한 번도 예측하지 못했고 확률 점수 자체도 구분력이 거의 없었다(AUC≈0.52)

검증 구간에서 high_vol 클래스의 비중이 매우 낮아(8/154)     
기본 Logistic Regression은 보수적인 예측을 수행하며 high_vol을 탐지하지 못했다.

ROC AUC 또한 0.52 수준으로 나타나 확률 구분력도 제한적이었다.

이에 따라 이후 단계에서는 스케일링 적용, 클래스 불균형 대응, 임계값 튜닝을 통해 ‘고변동성 탐지’ 목적에 맞는 평가(Recall 중심)로 모델을 재설계한다.

## StandardScaler

```python
# Scaler + balanced Logistic 파이프라인

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

pipe_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LogisticRegression(class_weight='balanced',
                              max_iter=1000,
                              random_state=42))
])

pipe_lr.fit(X_train, y_train)
```

```python
# 기본 threshold(0.5) 기준 성능 확인
# - class_weight='balanced'를 적용한 상태에서
# - 고변동성(high_vol) 구간을 얼마나 놓치지 않고 탐지하는지(recall)를 우선적으로 평가
# - precision은 낮더라도 early warning 관점에서는 허용

from sklearn.metrics import confusion_matrix, classification_report

y_pred = pipe_lr.predict(X_valid)

print("== Logistic Regression | threshold 0.50 ==")
print(confusion_matrix(y_valid, y_pred))
print(classification_report(y_valid, y_pred, digits=3))
```

## Logistic Regression 회귀 계수 확인

이 결과는 threshold 조정 및 시차 변수(lag) 추가를 통해 추가적인 성능 개선 여지가 있음을 시사한다.

## threshold 튜닝

```python
# 확률 예측
y_proba = pipe_lr.predict_proba(X_valid)[:, 1]
```

```python
# threshold별 성능 확인
# - high_vol은 소수 클래스이므로 accuracy 대신 precision/recall/F1을 확인
# - threshold를 낮추면 high_vol 탐지(Recall)는 증가하지만 오탐(FP)도 함께 증가(Precision 하락)
# - 운영 목적(탐지 우선 vs 경고 품질 우선)에 따라 threshold를 선택

from sklearn.metrics import precision_score, recall_score, f1_score

thresholds = np.arange(0.1, 0.6, 0.05)

results = []

for t in thresholds:
    y_pred_t = (y_proba >= t).astype(int)

    results.append({
        'threshold': t,
        'precision': precision_score(y_valid, y_pred_t, zero_division=0),
        'recall': recall_score(y_valid, y_pred_t),
        'f1': f1_score(y_valid, y_pred_t)
    })

results_df = pd.DataFrame(results)
results_df
```

```python
# 시각화 확인

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(results_df['threshold'], results_df['recall'], label='Recall')
plt.plot(results_df['threshold'], results_df['f1'], label='F1')
plt.plot(results_df['threshold'], results_df['precision'], label='Precision')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```

**결과:**    
threshold를 0.10~0.55 범위에서 조정하며 precision/recall/F1 변화를 확인하였다.     
검증 구간에서 high_vol 표본 수가 8개로 적어, threshold 변화에 따라 recall이 1.00에서 0.75로 계단 형태로 감소하는 패턴이 관측되었다.

낮은 threshold(0.10~0.30)에서는 모든 high_vol을 탐지하여 recall=1.00을 달성했지만, false positive가 증가하여 precision과 F1이 낮게 나타났다.
반대로 threshold를 높일수록 precision이 상승하면서 F1이 개선되었고, threshold 0.50 ~ 0.55 구간에서 F1이 가장 높게 나타났다.

따라서 운영 목적이 “놓치지 않는 경고(Recall 우선)”인지 “경고 품질(F1 우선)”인지에 따라 적절한 threshold를 선택할 필요가 있다.

```python
# threshold 간격 0.01로 수정하여 최적점 찾아보기

y_proba = pipe_lr.predict_proba(X_valid)[:, 1]

thresholds = np.arange(0.05, 0.95, 0.01)
rows = []
for t in thresholds:
    y_pred_t = (y_proba >= t).astype(int)
    rows.append((t,
                 precision_score(y_valid, y_pred_t, zero_division=0),
                 recall_score(y_valid, y_pred_t),
                 f1_score(y_valid, y_pred_t)))
results_df = pd.DataFrame(rows, columns=['threshold','precision','recall','f1'])

# F1 최대 threshold
results_df.loc[results_df['f1'].idxmax()]
```

**결과:**    
threshold를 0.01 간격으로 탐색한 결과, F1-score는 threshold 0.91에서 최대가 되었으나    
해당 구간은 recall이 0.25로 매우 낮아 실제 고변동성(high_vol)의 대부분을 놓치는 문제를 확인하였다.

이는 불균형 데이터에서 F1 최대화가 “경고를 거의 내지 않는 보수적 분류”로 수렴할 수 있음을 의미한다.    
따라서 본 프로젝트에서는 F1 최대 기준만으로 threshold를 선택하지 않고 고변동성 탐지 목적에 맞게     
recall을 일정 수준 이상 유지하는 제약을 둔 뒤 그 범위 내에서 precision 또는 F1이 최대가 되는 threshold를 선택한다.

```python
# 탐지(Early Warning) 목적: high_vol을 놓치지 않는 것을 최우선으로 함
# 1) recall=1.0을 만족하는 threshold 후보군을 먼저 필터링
# 2) 그 범위 내에서 precision이 최대가 되는 threshold를 선택 (오탐 최소화)

target_recall = 1.0

cand = results_df[results_df['recall'] >= target_recall]
cand.sort_values(['precision', 'f1'], ascending=False).head(10)
```

```python
# 대안 기준(보수적 운영): recall을 일정 수준 이상 유지하면서 경고 품질(F1)을 개선
# recall>=0.75 후보군 중 F1이 최대인 threshold를 확인

target_recall = 0.75

cand = results_df[results_df['recall'] >= target_recall]
cand.sort_values('f1', ascending=False).head(10)
```

**결과:**    
threshold 탐색 결과, recall=1.0을 유지하는 구간(0.21~0.30) 내에서 precision이 가장 높은 값은 threshold=0.30으로 확인되었다.    
해당 기준에서는 검증 구간의 고변동성(high_vol) 사례를 모두 탐지(recall=1.0)하면서도     
동일 조건 내에서 오탐 비율을 최소화(precision 최대화)할 수 있으므로 사전 경고 목적의 운영 threshold로 0.30을 채택하였다.

- threshold = 0.30
- precision = 0.095238
- recall = 1.00
- f1 = 0.173913

```python
# threshold = 0.30 의 confusion matrix

best_t = 0.30
y_pred_best = (y_proba >= best_t).astype(int)
cm = confusion_matrix(y_valid, y_pred_best)
cm
```

**결과:**    
threshold=0.30 기준에서 고변동성 사례는 모두 탐지되었으며(FN=0) 다만 불균형 데이터 특성상 오탐(FP)이 함께 증가하는 trade-off가 존재하였다.    
본 프로젝트는 “사전 경고” 목적이므로, 오탐 비용보다 놓침 비용을 더 크게 두고 threshold를 결정하였다.

## XGBoost

Logistic Regression을 통해 구조적 신호의 존재를 확인한 뒤 변수 간 비선형 관계와 상호작용을 학습하기 위해 트리 기반 부스팅 모델인 XGBoost를 적용하였다.

```python
# X / y 구성
# - lag 없이 현재 시점의 외생 변수만 사용
# - 시간 정보 누수를 방지하기 위해 Date 기준으로 train / valid 분리

X = df[feature_cols].copy()
y = df['high_vol']

split_mask = df['Date'] >= split_date

X_train = X.loc[~split_mask]
X_valid = X.loc[split_mask]
y_train = y.loc[X_train.index]
y_valid = y.loc[X_valid.index]

print("train rows:", X_train.shape)
print("valid rows:", X_valid.shape)
print("valid high_vol:", int(y_valid.sum()))
```

```python
# XGBoost 기본 모델 설정
# - 데이터 규모가 크지 않으므로 트리 깊이를 제한(max_depth=3)
# - 학습률을 낮춰 과적합을 방지(learning_rate=0.05)
# - subsample / colsample_bytree로 랜덤성 부여
# - scale_pos_weight로 클래스 불균형 보정

from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=200,
    max_depth=3,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='binary:logistic',
    eval_metric='logloss',
    scale_pos_weight=(len(y_train) - y_train.sum()) / y_train.sum(),
    random_state=42
)

xgb.fit(X_train, y_train)
```

```python
# 기본 threshold(0.5) 기준 성능 확인
# - 모델의 기본 분류 성향 파악 목적
# - 조기 경고 관점에서는 참고용 결과

y_pred_05 = xgb.predict(X_valid)

print("== XGBoost | threshold 0.50 ==")
print(confusion_matrix(y_valid, y_pred_05))
print(classification_report(y_valid, y_pred_05, digits=3, zero_division=0))
```

**결과:**    
기본 threshold(0.5) 기준에서 XGBoost 모델은 전체 정확도는 높게 나타났으나 변동성 확대(high_vol) 탐지 성능은 제한적인 모습을 보였다.

총 8건의 high_vol 중 1건만을 탐지하여 recall은 0.125 수준에 그쳤으며 이는 보수적인 분류 기준으로 인해 대부분의 변동성 확대 구간을 놓치고 있음을 의미한다.    
따라서 본 문제의 목적이 ‘변동성 확대 구간을 사전에 경고하는 탐지 모델’임을 고려할 때 threshold 조정을 통한 recall 개선이 필요하다고 판단하였다.

## threshold 튜닝

```python
# high_vol(1) 클래스에 대한 예측 확률 추출
# - threshold 조정을 위해 predict_proba 사용

y_proba = xgb.predict_proba(X_valid)[:, 1]
```

```python
# threshold를 0.05 ~ 0.95 범위에서 0.01 간격으로 변화시키며
# precision / recall / f1-score의 변화를 확인

thresholds = np.arange(0.05, 0.95, 0.01)

rows = []
for t in thresholds:
    y_pred_t = (y_proba >= t).astype(int)
    rows.append({
        'threshold': t,
        'precision': precision_score(y_valid, y_pred_t, zero_division=0),
        'recall': recall_score(y_valid, y_pred_t),
        'f1': f1_score(y_valid, y_pred_t)
    })

results_xgb = pd.DataFrame(rows)
results_xgb
```

**결과:**    
threshold 변화에 따른 성능을 확인한 결과 낮은 threshold 구간에서는 recall이 높게 유지되는 반면 precision은 상대적으로 낮게 나타났다.

반대로 threshold가 높아질수록 모델은 보수적으로 예측하게 되며 high_vol 탐지가 거의 이루어지지 않는 구간도 확인되었다.    
이는 변동성 확대 탐지 문제에서 threshold 선택이 모델 성능에 결정적인 영향을 미친다는 점을 보여준다.

```python
# recall = 1.0을 유지하면서 precision이 최대가 되는 threshold 탐색

cand = results_xgb[results_xgb['recall'] == 1.0] \
    .sort_values('precision', ascending=False)

cand.head(10)
```

```python
# recall ≥ 0.75 기준에서의 최적 threshold 선택
# # f1-score가 가장 높은 threshold 후보 확인

results_xgb[results_xgb['recall'] >= 0.75] \
    .sort_values('f1', ascending=False) \
    .head(10)
```

**결과:**    
XGBoost 모델은 기본 threshold 기준에서는 변동성 확대(high_vol) 탐지 성능이 제한적이었으나 threshold 조정을 통해 recall을 유의미하게 개선할 수 있었다.

다만 본 데이터의 규모와 클래스 분포 특성상 recall을 1.0으로 완전히 유지하는 구간은 존재하지 않았으며    
이에 따라 일정 수준 이상의 recall을 확보하는 현실적인 탐지 기준을 선택하였다.

이러한 결과는 변동성 확대 탐지가 단순한 분류 문제가 아닌 확률 기반 의사결정 문제임을 시사하며     
모델의 예측 확률을 어떻게 해석하고 활용할 것인지가 모델 자체의 성능만큼 중요함을 보여준다.

recall ≥ 0.75 조건 하에서 threshold를 비교한 결과 threshold 약 0.13~0.14 구간에서 f1-score가 가장 높게 나타났다.    
해당 구간은 전체 변동성 확대 사례의 약 75%를 탐지하면서도 과도한 오탐(false positive)을 일정 수준으로 제어하는 균형적인 지점으로 해석할 수 있다.

## Logistic vs XGBoost threshold 비교 요약

Logistic Regression은  
변동성 확대 탐지 문제에서 구조적 신호의 존재를 확인하는  
기준 모델(baseline)로서 의미 있는 역할을 수행하였다.

특히 threshold 조정을 통해  
recall을 안정적으로 확보할 수 있었으며,  
각 변수의 방향성과 영향력을 해석하기에 적합한 모델이었다.

반면 XGBoost는  
변수 간 비선형 관계와 복합적인 상호작용을 학습할 수 있는 모델로,  
threshold 조정을 통해 탐지 성능을 유연하게 제어할 수 있었다.

다만 본 데이터 규모에서는  
recall을 완벽히 유지하는 구간은 존재하지 않았으며,  
이에 따라 일정 수준 이상의 recall을 확보하는  
현실적인 탐지 기준을 선택하는 접근이 필요했다.

종합적으로 볼 때,  
Logistic Regression은 구조적 이해와 기준 설정에 강점이 있었고,  
XGBoost는 비선형 패턴을 반영한  
확률 기반 조기 경고 모델로서의 가능성을 보여주었다.

## Linear Regression (베이스라인)

- 외생 변수와 변동성 수준 간의 선형 관계만을 가정
- 비선형 모델과의 성능 비교를 위한 기준선 역할

```python
# 모델 학습 Linear Regression

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_valid)

lr_mae = mean_absolute_error(y_valid, y_pred_lr)
lr_rmse = np.sqrt(mean_squared_error(y_valid, y_pred_lr))
lr_r2 = r2_score(y_valid, y_pred_lr)

print("Linear Regression performance")
print("MAE :", round(lr_mae, 4))
print("RMSE:", round(lr_rmse, 4))
print("R2  :", round(lr_r2, 4))
```

**결과:**    
Linear Regression은 외생 변수와 WTI 변동성 수준(wti_vol_4w) 간의 선형 관계를 가정한 베이스라인 회귀 모델로 사용하였다.    

모델 평가 결과, MAE와 RMSE 기준의 절대 오차는 비교적 작은 수준을 보였으나     
결정계수(R²)가 음수로 나타나 변동성 수준에 대한 설명력은 매우 제한적인 것으로 확인되었다.     
이는 해당 모델이 단순히 평균값으로 예측하는 경우보다도 변동성의 변화를 잘 설명하지 못함을 의미한다.

이러한 결과는 WTI 변동성 수준이 개별 외생 변수의 선형 결합보다는     
변수 간의 비선형적 상호작용이나 특정 조건에서의 임계 효과에 의해 형성될 가능성이 높다는 점을 시사한다.    
따라서 이후 분석에서는 비선형 관계와 변수 간 상호작용을 효과적으로 학습할 수 있는     
트리 기반 회귀 모델(XGBoost Regressor)을 적용하여 변동성 수준 예측 성능을 추가적으로 검증한다.

## XGBoost Regressor

```python
# XGBoost Regressor
# - 선형 회귀에서 포착되지 않은 비선형 관계 및 변수 간 상호작용 학습
# - 변동성 수준(wti_vol_4w)에 대한 설명력 개선 여부 확인 목적

from xgboost import XGBRegressor

xgb_reg = XGBRegressor(
    n_estimators=300,
    max_depth=3,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='reg:squarederror',
    random_state=42
)

xgb_reg.fit(X_train, y_train)

y_pred_xgb = xgb_reg.predict(X_valid)
```

```python
# 회귀 성능 평가
# - MAE / RMSE: 예측 오차의 절대적 크기
# - R²: 변동성 수준에 대한 설명력

mae = mean_absolute_error(y_valid, y_pred_xgb)
rmse = np.sqrt(mean_squared_error(y_valid, y_pred_xgb))
r2 = r2_score(y_valid, y_pred_xgb)

print("== XGBoost Regressor performance ==")
print("MAE :", round(mae, 4))
print("RMSE:", round(rmse, 4))
print("R2  :", round(r2, 4))
```

```python
# Linear Regression과 XGBoost Regressor 성능 비교
# - 동일한 입력 변수 기준에서 모델 구조 차이에 따른 성능 변화 확인

print("== Model Comparison ==")
print(f"Linear  | MAE: {round(lr_mae,4)} | RMSE: {round(lr_rmse,4)} | R2: {round(lr_r2,4)}")
print(f"XGBoost | MAE: {round(mae,4)} | RMSE: {round(rmse,4)} | R2: {round(r2,4)}")
```

**결과:**    
비선형 관계를 학습할 수 있는 XGBoost Regressor를 적용하였으나     
본 데이터 규모와 관측 기간이 제한적인 상황에서는 회귀 문제에서도 설명력(R²) 개선이 뚜렷하게 나타나지 않았다.

이는 변동성 수준 예측 문제가 단순한 지도 학습 회귀 문제라기보다는    
구조적 변화, regime 전환, 이벤트성 충격의 영향을 크게 받는 복합적인 시계열 문제임을 시사한다.

반면 Part 2의 분류 모델에서는 변동성 확대 여부(high_vol)를 탐지하는 데 있어 의미 있는 성능을 확보할 수 있었으며     
이는 변동성 수준 자체를 정밀하게 예측하기보다는 위험 구간을 사전에 감지하는 접근이 본 문제에 보다 현실적인 활용 방식임을 뒷받침한다.

따라서 본 분석에서는 변동성 수준의 정확한 회귀 예측보다는 확률 기반 탐지 및 조기 경고 모델링이 WTI 시장 분석에 더 적합한 접근임을 결론으로 도출한다.

# 최종 결론

# 최종 결론 (WTI 원유 가격 리스크 모니터링 및 예측 시스템)

본 프로젝트는 **WTI 주간 유가를 단순히 예측하는 수준을 넘어**,  
달러지수·금리·재고·주식시장 변동성·OPEC 이벤트와 같은 외생 변수를 결합하여  
**‘변동성 확대(high_vol) 구간을 사전에 경고(Early Warning)하는 리스크 모니터링 모델’**을 구축하는 것을 목표로 했다.

---

## 1) 탐색 분석(Part 1) 결론: 유가 변동성은 ‘주기’가 아니라 ‘국면 + 충격’의 문제

TS-EDA 결과, WTI 가격은 뚜렷한 계절성이나 반복 주기보다  
**특정 시점의 급등·급락이 전체 변동성을 지배하는 구조**가 강하게 관찰되었다.  

따라서 본 문제는 “가격 레벨을 정확히 맞히는 예측”보다  
**변동성이 급격히 커지는 위험 국면을 먼저 포착하는 탐지 문제**로 정의하는 것이 더 현실적이라고 판단하였다.

또한 외생 변수와의 관계는 전 기간에 걸쳐 단순 상관으로 설명되기 어려웠으며,  
특히 달러 수익률, 금리 변화 등 **레벨/변화량의 단순 관계는 일관되지 않았다.**  

반면 다음과 같은 변수들은 “레벨”보다 **변동성/충격 형태**로 구성되었을 때  
WTI 고변동성(high_vol) 구간과 더 구조적인 연결 가능성을 보였다.

- **S&P500 변동성(sp500_vol_4w)**  
- **금리 변동성(rate_vol_4w)**  
- **재고 충격(inventory_change / inventory_vol_4w)**  
- **정책 이벤트(OPEC_ProdDirection / OpecEventDummy)**  

즉, 유가 리스크는 단일 요인보다는 금융·수급·정책 신호가 결합되는 국면에서 강화될 가능성이 높다는 관찰을 얻었다.

---

## 2) 분류 모델(Part 2) 결론: ‘리스크 경고’는 분류 + threshold 운영이 핵심

본 프로젝트의 핵심 목표는 **고변동성(high_vol) 탐지**이며,  
이는 전형적인 **소수 클래스 탐지 문제(검증 구간 high_vol 8/154)** 로 나타났다.

---

### 2-1. Logistic Regression 결론: 구조적 신호 존재를 확인한 기준 모델

- 기본 Logistic Regression은 불균형으로 인해 **high_vol(1)을 한 번도 예측하지 못했고(AUC≈0.52)**  
  사실상 경고 기능이 작동하지 않았다.
- 그러나 **StandardScaler + class_weight='balanced' 적용 후**,  
  lag 변수가 없는 상태에서도 **recall 0.75 수준**을 확보하며  
  **외생 변수 조합만으로도 ‘탐지 가능한 신호’가 존재함을 확인**하였다.

또한 계수 해석을 통해

- **sp500_vol_4w(주식시장 변동성)**
- **wti_return(원유 자체 충격)**
- **inventory_change(수급 충격)**
- **OPEC_ProdDirection(정책 방향 이벤트)**  

등이 high_vol 확률을 올리는 방향으로 작동함을 확인하였다.  
이는 유가 변동성 확대가 금융시장 불안과 수급·정책 신호에 의해 강화될 수 있음을 뒷받침한다.

결과적으로 Logistic Regression은 성능 자체뿐 아니라  
**변수의 방향성과 영향력을 해석할 수 있는 기준선(baseline)** 으로서 의미를 갖는다.

---

### 2-2. Threshold 튜닝 결론: F1 최대가 아니라 ‘운영 목적’이 기준

불균형 탐지 문제에서는 F1 최대화가  
**경고를 거의 내지 않는 보수적 분류(낮은 recall)** 로 수렴할 수 있음을 확인하였다.  

따라서 본 프로젝트에서는 “사전 경고 시스템”이라는 목적에 맞추어

- **놓침 비용(FN)을 크게**
- **오탐 비용(FP)을 상대적으로 작게**

두는 의사결정 구조로 threshold를 선택하였다.

- Logistic Regression은 **recall=1.0 제약 하에서 precision이 최대인 threshold=0.30**을 운영 후보로 채택하였다.  
  이 기준은 **high_vol을 놓치지 않는(FN=0)** 대신 오탐(FP)이 증가하는 trade-off를 명시적으로 수용한다.

---

### 2-3. XGBoost 결론: 비선형/상호작용을 반영한 확률 기반 경고 모델

XGBoost는 기본 threshold(0.5)에서는  
high_vol 탐지 성능이 제한적이었으나(recall 0.125),  

threshold 조정을 통해  
**recall을 0.75 수준까지 확보하는 운영 지점(약 0.13~0.14)** 을 찾을 수 있었다.

feature importance 및 SHAP 분석 결과, high_vol 예측은

- **S&P500 변동성(sp500_vol_4w)** 이 가장 큰 핵심 요인
- **WTI 자체 충격(wti_return)** 이 다음으로 중요한 요인
- **금리/달러 변동성(rate_vol_4w / usd_vol_4w)** 은 금융 환경 불안정성을 반영하는 보조 요인
- **OPEC_ProdDirection** 은 특정 시점에서 예측을 강하게 밀어 올리는 트리거 요인

으로 반복적으로 확인되었다.

즉, 모델은 단일 이벤트 여부보다  
**“금융시장 변동성 축적 + 원유 가격 충격 + (일부 시점) 정책 트리거”**의 결합 구조를  
고변동성 경고(high_vol)로 식별하는 경향을 보였다.

---

## 3) 회귀 모델(Part 3) 결론: 변동성 ‘수준’ 예측은 데이터 조건상 비현실적

Part 3에서는 변동성 수준(wti_vol_4w)을 회귀로 예측했으나,  
rolling 기반 타깃/피처 생성 과정에서 표본이 크게 감소하여  
(train 17, valid 6 수준) 모델 평가가 매우 불안정해졌다.

그 결과 Linear Regression과 XGBoost Regressor 모두 **R²가 음수**로 나타나  
변동성 수준을 지도학습 회귀로 설명하기 어려움을 확인하였다.

이는 변동성 수준이

- 구조적 변화(regime shift)
- 이벤트성 충격
- 비정상성(Non-stationarity)

의 영향을 크게 받기 때문이며,  
현재 데이터 규모/구성에서는 **정밀한 회귀 예측보다 위험 구간 탐지가 훨씬 실용적**이라는 결론으로 이어졌다.
